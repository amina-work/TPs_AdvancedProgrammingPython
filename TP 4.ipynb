<b>Exercice 1 :</b> Créez un programme qui demande à l'utilisateur le nombre de fois qu'une pièce sera lancée et qui affiche pile ou face (aléatoire) ainsi que le nombre de fois où l'on obtient pile et le nombre de fois où l'on obtient face.
**<center>Exemple d'exécution</center>**
<p style="padding: 10px; border: 1px solid black;">
JEU PILE OU FACE
<b style="color:blue;"><br>Donnez le nombre de lancers : 5</b>
<br>Lancé 1 : Pile
<br>Lancé 2 : Pile
<br>Lancé 3 : Face
<br>Lancé 4 : Pile
<br>Lancé 5 : Face
<br>Nombre piles : 3
<br>Nombre faces : 2
</p>




````python
#My code
import random

L = ['Pile', 'Face']
print("JEU PILE OU FACE")
n = int(input("Donnez le nombre de lancers: "))
lst = []
p = 0
f = 0
for i in range(n+1):
        lst.append(i)
for i in range(1, len(lst)):
    res = random.choice(L)
    print("Lancé", i, ":", res)
    if res == L[0]:
        p += 1
    elif res == L[1]:
        f += 1
print("Nombre piles: ", p)
print("Nombre faces: ", f)
````

<b>Exercice 2 :</b> On voudrait créer un programme qui calcule la taille moyenne des joueurs d'une équipe de basketball. Le programme demande d'abord à l'utilisateur le nombre de joueurs ainsi que leurs tailles (en centimètres). Vous devez vérifier que les valeurs données sont positives (validation des données). Le programme affichera la taille moyenne de joueurs de l'équipe.
**<center>Exemple d'exécution</center>**
<p style="padding: 10px; border: 1px solid black;">
CALCULATEUR DE LA TAILLE MOYENNE
<b style="color:blue;"><br>Donnez le nombre de joueurs : -5
<br>Le nombre de joueurs ne peut pas être négatif. Essayez à nouveau : 5
<br>Taille joueur 1 : 182
<br>Taille joueur 2 : 0
<br>Taille incorrecte. Donnez Taille joueur 2 : 178
<br>Taille joueur 3 : 185
<br>Taille joueur 4 : 173
<br>Taille joueur 5 : 188</b>
<br>La taille moyenne des joueurs  : 181.20
</p>




````python
#My code
print("CALCULATEUR DE LA TAILLE MOYENNE")
moy = 0
n = int(input("Donnez le nombre de joueurs : "))
while n <= 0:
    n = int(input("Le nombre de joueurs ne peut pas etre négatif ou null, Essayer à nouveau : "))
lst = []
for i in range(n+1):
    lst.append(i)
for i in range(1, len(lst)):
    taille = int(input("Taille joueur " + str(i) + " : "))
    while taille <= 0:
        taille = int(input("Taille incorrecte. Donnez Taille joueur " + str(i) + " : "))
    moy += taille
moy /= n
print("La taille moyenne des joueurs :", format(moy, ".2f"))
````

<b>Exercice 3 :</b> Créer un programme qui demande à l'utilisateur de saisir les notes d'étudiants jusqu'à ce que l'utilisateur entre la valeur -1. Une note doit être comprise entre 0 et 20. Le programme affichera la note maximale ainsi que la moyenne des notes.
**<center>Exemple d'exécution</center>**
<p style="padding: 10px; border: 1px solid black;">
TROUVER LE MAXIMUM
<b style="color:blue;"><br>Donnez la note 1 : 15
<br>Donnez la note 2 : -8
<br>La note est incorrecte. Donnez la note 2 : 12
<br>Donnez la note 3 : 10
<br>Donnez la note 4 : 18
<br>Donnez la note 5 : -1</b>
<br>La note maximale est : 13.75
<br>La moyenne est : 18
</p>

````python
#My code
print("TROUVER LE MAXIMUM")
moy = 0
mx = 0
i = 1
while True:
    n = int(input("Donnez la note " +str(i)+ " :"))
    if n == -1:
        break #breaking out of the loop when value is -1
    if n >= mx: #getting the maximum value
        mx = n
    while n < 0:
        n = int(input("La note est incorrect. Donnez la note " +str(i)+ " :"))
    while n > 20:
        n = int(input("La note est incorrect. Donnez la note " +str(i)+ " :"))
    i += 1
    moy += n
moy = moy / (i-1)
print("La note maximale est:", mx)
print("La moyenne est:", moy)
````

<b>Exercice 5 :</b> Implémentez un programme qui affiche les solutions du jeu FizzBuzz. Le programme affiche tous les chiffres de 1 jusqu'à 100.
Si le nombre est divisible par 3 (par exemple 3 ou 12), le programme affiche "Fizz" au lieu du chiffre.
Si le nombre est divisible par 5 (par exemple (5 ou 20), le programme affiche "Buzz" au lieu du chiffre.
Si le nombre est divisible par 3 et 5 (par exemple 15 ou 30), le programme affiche "FizzBuzz" au lieu du chiffre.
**<center>Exemple d'exécution</center>**
<p style="padding: 10px; border: 1px solid black;">
1
<br>2
<br>Fizz
<br>4
<br>Buzz
<br>Fizz
<br>7
<br>8
<br>Fizz
<br>Buzz
<br>11
<br>Fizz
<br>13
<br>14
<br>FizzBuzz
<br>...
</p>

````python
#My code
for i in range(1, 101, 1):
    if i % 3 == 0 and i % 5 == 0:
        print("FizzBuzz")
    elif i % 3 == 0:
        print("Fizz")
    elif i % 5 == 0:
        print("Buzz")
    else:
        print(i)
````

<b>Exercice 6 :</b> Complétez le code ci-dessous qui permet de trouver et d'afficher la valeur minimale de la liste en utilisant la boucle for puis en utilisant la boucle while.
**<center>Exemple d'exécution</center>**
<p style="padding: 10px; border: 1px solid black;">
Le valeur minimale est : 3
</p>

````python
#complétez le code (boucle for)
mn = None
for value in [17, 21, 5, 18, 29, 3] :
    if mn is None or value < mn:
        mn = value
print("La valeur minimale est:", mn)
````

````python
#complétez le code (boucle while)
mn = None
lst = [17, 21, 5, 18, 29, 3] 
value = 0
while value < len(lst):
    if mn is None or lst[value] < mn:
        mn = lst[value]
    value += 1
print("La valeur minimale est:", mn)
````
